Loop Statement

```python
week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for x in week:
  print(x)
```

break Statements (Loop Statement):

```python
week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for x in week:
  print(x)
  if x == "Thursday":
    break
```

```python
week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for x in week:
  if x == "Thursday":
    print(x)
    break
```

```python
week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for x in week:
  if x == "Thursday":
    break
  print(x)
```

Loping through a string

```python
for x in "Normil":
  print(x)
```

Nested Loops

```python
adjective = ['red', 'big', 'tasty']
fruits = ['apple', 'banana', 'cherry']

for x in adjective:
  for y in fruits:
    print(x,y)
```

While Loop

```python
i = 1
while i <= 6:
  print(i)
  i += 1 # i = i + 1 Assignment operator
```

break statement (While Loop):


```python
i = 1
while i <= 6:
  print(i)
  if i ==3:
    break
  i += 1 # stop from being infinite
```

```python
i = 1
while i <= 6:
  if i ==3:
    break
  print(i)
  i += 1 # stop from being infinite
```

```python
i = 1
while i <= 6:
  if i ==3:
    print(i)
    break
  i += 1 # stop from being infinite
```

The Else statement

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print('i is no longer less than 6')
```
